```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Total Schools
school_count = len(school_data_complete["school_name"].unique())
#Total Students
student_count = len(school_data_complete["student_name"])
#Total Budget
total_budget = sum(school_data["budget"])
#Average Maths Score
avg_maths = school_data_complete["maths_score"].mean()
#Average Reading Score
avg_reading = school_data_complete["reading_score"].mean()

#Percentage of students with passing maths score
pass_mathsscore_df = school_data_complete.loc[school_data_complete["maths_score"] >= 50, [
    "maths_score"]]
pass_mathsscore_df.head()
percent_maths_pass = len(pass_mathsscore_df["maths_score"]) / student_count * 100

#Percentage of students with passing reading score
pass_readscore_df = school_data_complete.loc[school_data_complete["reading_score"] >= 50, [
    "reading_score"]]
pass_readscore_df.head()
percent_read_pass = len(pass_readscore_df["reading_score"]) / student_count * 100

#Percentage of students with overall passing in maths and reading score
overall_passing_df = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50) , 
["student_name","maths_score","reading_score"]]
overall_passing_df.head()
overall_passing = len(overall_passing_df["student_name"]) / student_count * 100

#Local_govt_area Summary Dataframe
area_summary = pd.DataFrame({"Total Schools": [school_count] ,
                                   "Total Students": student_count ,
                                   "Total Budget": total_budget,
                                   "Average Maths Score": avg_maths,
                                   "Average Reading Score": avg_reading,
                                   "% Passing Maths": percent_maths_pass,
                                   "% Passing Reading": percent_read_pass,
                                   "% Overall Passing": overall_passing})
#Formatting the displayed data
format_dict = {'Total Students':'{:,.0f}', 'Total Budget': '${:,.2f}'}
area_summary.head().style.format(format_dict)
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Get Unique School Names
grouped_lg_df = school_data_complete.groupby(["school_name"])
#Total no of students in each school
lg_no_of_students =  grouped_lg_df["size"].first()
#School type
lg_school_type = grouped_lg_df["type"].first()
#Total budget for each school
lg_total_budget = grouped_lg_df["budget"].first()
#Per Student budget per school
perStudent_budget = lg_total_budget / lg_no_of_students
#Average Maths and Reading Score for each school
avg_maths_score = grouped_lg_df["maths_score"].mean()
avg_reading_score = grouped_lg_df["reading_score"].mean()

#Percentage of Passing Maths Score for each school
pass_mathsscorepersch_df = school_data_complete.loc[school_data_complete["maths_score"] >= 50, [
    "maths_score","school_name"]]
grouped_onlypassMaths = pass_mathsscorepersch_df.groupby("school_name")
percent_maths_pass_perschool = grouped_onlypassMaths["maths_score"].size() / lg_no_of_students * 100

#Percentage of Passing Reading score for each school
pass_readscorepersch_df = school_data_complete.loc[school_data_complete["reading_score"] >= 50, [
    "reading_score","school_name"]]
grouped_onlypassread = pass_readscorepersch_df.groupby("school_name")
percent_read_pass_perschool = grouped_onlypassread["reading_score"].size() / lg_no_of_students * 100

#Percentage of Overall Passing in both Maths and Reading for each school
overall_passingpersch_df = school_data_complete.loc[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50) , 
["maths_score","reading_score","school_name"]]
grouped_overallpass_perschool = overall_passingpersch_df.groupby("school_name")
overall_passing_perschool = grouped_overallpass_perschool["school_name"].size() / lg_no_of_students * 100

#School Summary Dataframe
per_school_summary = pd.DataFrame({"School Type": lg_school_type,
                              "Total Students": lg_no_of_students,
                              "Total School Budget": lg_total_budget,
                              "Per Student Budget":perStudent_budget,
                              "Average Maths Score" :avg_maths_score,
                              "Average Reading Score" :avg_reading_score,
                              "% Passing Maths": percent_maths_pass_perschool,
                              "% Passing Reading": percent_read_pass_perschool,
                              "% Overall Passing": overall_passing_perschool
                              })

#Formatting the displayed data
format_dict_perschool = {'Total School Budget':'${:,.2f}', 'Per Student Budget': '${:,.2f}'}
per_school_summary.head(school_count).style.format(format_dict_perschool)
```

## Maths Scores by Year

```python
#Average maths score for Year 9 
year_9_df = school_data_complete.loc[school_data_complete["year"] == 9, [
    "maths_score","school_name"]]
grouped_year9 = year_9_df.groupby("school_name")

avgmathsscore_year9 = grouped_year9["maths_score"].mean()

#Average maths score for Year 10 
year_10_df = school_data_complete.loc[school_data_complete["year"] == 10, [
    "maths_score","school_name"]]
grouped_year10 = year_10_df.groupby("school_name")

avgmathsscore_year10 = grouped_year10["maths_score"].mean()

#Average maths score for Year 11
year_11_df = school_data_complete.loc[school_data_complete["year"] == 11, [
    "maths_score","school_name"]]
grouped_year11 = year_11_df.groupby("school_name")

avgmathsscore_year11 = grouped_year11["maths_score"].mean()

#Average maths score for Year 12 
year_12_df = school_data_complete.loc[school_data_complete["year"] == 12, [
    "maths_score","school_name"]]
grouped_year12 = year_12_df.groupby("school_name")

avgmathsscore_year12 = grouped_year12["maths_score"].mean()

#Dataframe of Average maths score for Year 9,10,11,12
maths_scores_by_year = pd.DataFrame({"Year 9" : avgmathsscore_year9,
                                         "Year 10" : avgmathsscore_year10,
                                         "Year 11" : avgmathsscore_year11,
                                         "Year 12" : avgmathsscore_year12
                                        })

maths_scores_by_year.head(school_count)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python

# Create bins to group school spending
spending_bins = [0, 585, 630, 645, 680]

# Create labels for these bins
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending_df = per_school_summary[["Per Student Budget","Average Maths Score","Average Reading Score","% Passing Maths", "% Passing Reading", "% Overall Passing"]]
# Slice the data and place it into bins and Place the data series into a new column inside of the DataFrame
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labels)
school_spending_df.head()

# Calculate Average with in GroupBy object based upon "Spending Ranges(Per Student)"

spending_maths_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"].round(2)
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"].round(2)
spending_passing_maths = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"].round(2)
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"].round(2)
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"].round(2)

#Create Dataframe to diaplay Spending Summary
spending_summary = pd.DataFrame({ "Average Maths Score" : spending_maths_scores,
                                 "Average Reading Score" : spending_reading_scores,
                                 "% Passing Maths" : spending_passing_maths,
                                 "% Passing Reading" : spending_passing_reading,
                                 "% Overall Passing" : overall_passing_spending,

})
spending_summary.head()
```
